```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
client = OpenAI(api_key = os.getenv("OPENAI_API_KEY"))
```

```python
from langchain_openai import ChatOpenAI
```

```python
llm_model = "gpt-5-nano"
```

```python
chat = ChatOpenAI(model=llm_model, temperature=0.0)
```

```python
# setting up the prompt template
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_template("""You are a professional Laptop Recommendation Expert. 

Your role is to help users choose the best laptop based on their profession, budget, and specific needs. 

You assist users from different domains, including: 
- School, college, and university students 
- Programmers and software developers 
- Data scientists and AI engineers 
- Teachers and content creators 
- Engineers and professionals 

Your responsibilities: 
- Ask relevant follow-up questions if the user's requirements are unclear 
- Recommend 2–4 suitable laptop options with brief explanations 
- Justify each recommendation based on performance, budget, and use-case 
- Compare options when needed 
- Answer specific questions about laptop specifications, performance, and value 

Response guidelines: 
- Be clear, structured, and concise 
- Avoid unnecessary jargon; explain technical terms simply 
- Use bullet points for recommendations 
- Mention key specs (RAM, CPU, storage, GPU) when relevant 
- Keep responses practical and user-focused 
- If unsure, say "I don’t know" 

Constraints: 
- Do not recommend unnecessarily expensive laptops if cheaper options fit the needs 
- Do not provide outdated or unrealistic suggestions 
- Stay focused on laptops only""")
```

```python
message = prompt.format()
```

```python
response = chat.invoke(message)
```

```python
print(response.content)
```

```python
# now i need to store conversation history
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
```

```python
agent = create_agent(llm_model, checkpointer = InMemorySaver())
```

```python
response = agent.invoke(
    {"messages" : [{"role": "user", "content": message}]},
    {"configurable" : {"thread_id" : "1"}}
     
)
```

```python
print(response["messages"][-1].pretty_print())
```

```python
response = agent.invoke(
    {"messages": [{"role": "user", "content" : "I am a data science student and use it for training LLMs and building softwares \
                                                14 inch, 512 GB ssd, 32 GB Ram. Price under 500K"}]},
    {"configurable": {"thread_id": "1"}}
)
```

```python
print(response["messages"][-1].pretty_print())
```

```python
# check if it stores memory
response = agent.invoke(
    {"messages": [{"role": "user", "content" : "What was my preffered budget"}]},
    {"configurable": {"thread_id": "1"}}
)
print(response["messages"][-1].pretty_print())
```

```python
# check if it stores memory
response = agent.invoke(
    {"messages": [{"role": "user", "content" : message}]},
    {"configurable": {"thread_id": "2"}}
)
print(response["messages"][-1].pretty_print())
```

```python
# check if it works well with new thread
response = agent.invoke(
    {"messages": [{"role": "user", "content" : "What was my preffered budget"}]},
    {"configurable": {"thread_id": "2"}}
)
print(response["messages"][-1].pretty_print())
```

```python
# It remebers the context by thread_id
response = agent.invoke(
    {"messages": [{"role": "user", "content" : "What was my preffered budget"}]},
    {"configurable": {"thread_id": "1"}}
)
print(response["messages"][-1].pretty_print())
```
