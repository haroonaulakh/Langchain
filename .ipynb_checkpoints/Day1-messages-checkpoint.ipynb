```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from langchain_openai import ChatOpenAI
```

```python
load_dotenv()
apikey = os.getenv("OPENAI_API_KEY")
```

```python
client = OpenAI(api_key=apikey)
```

```python
# Simple chat prompt
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
model_chat = ChatOpenAI(model="gpt-5-nano")
```

```python
prompt = ChatPromptTemplate.from_template("Write 5 lines about {topic}. Give numbering also.")
```

```python
chain = prompt | model_chat
```

```python
print(chain.invoke({"topic":"recursion"}).content)
```

AI Assitant

```python
from langchain.messages import HumanMessage, SystemMessage, AIMessage
from langchain.agents import create_agent
```

```python
sys_msg = SystemMessage("You are a helpful assistant who teaches maths.")
hum_msg = HumanMessage("Can You help me with polynomial functions.")
```

```python
messages = [sys_msg, hum_msg]
```

```python
response = model_chat.invoke(messages)
```

```python
print(response.content)
```

```python
print(response.usage_metadata)
```
